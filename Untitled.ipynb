```python
# example for understanding 

import pandas as pd

# example dataset 
df = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male'],
    'Education_Level': ['High School', 'Graduate', 'Graduate', 'PhD', 'High School', 'PhD', 'Graduate', 'Graduate'],
    'Marital_Status': ['Single', 'Married', 'Married', 'Single', 'Single', 'Married', 'Divorced', 'Single'],
    'Churn': [0, 1, 0, 1, 0, 1, 1, 0]  # 0 = stay, 1 = gone
})

print("ðŸ”¹ UrsprÃ¼nglicher DataFrame:")
print(df)
```

```python
def encoder_helper(df, category_lst, response):
    if not pd.api.types.is_numeric_dtype(df[response]):
        raise ValueError(f"Error")

    for column in category_lst:
        if column not in df.columns:
            continue

        mean_churn = df.groupby(column)[response].mean()
        df[column + '_' + response] = df[column].map(mean_churn)

    return df
```

```python
category_lst = ['Gender', 'Education_Level', 'Marital_Status']

df_encoded = encoder_helper(df.copy(), category_lst, 'Churn')

print("\nâœ… DataFrame nach dem Encoding:")
print(df_encoded)
```
